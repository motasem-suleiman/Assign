# Hypsometric curve

```python
import rasterio as rio
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dem_fname = 'rcew_dem_5m_.tif'
nbins = 35
```

```python
rcew_dem = rio.open(dem_fname)
rcew_dem
```

```python
rcew_dem.meta
```

```python
rcew_elev = rcew_dem.read(1)
rcew_elev
```

```python
rcew_elev = np.where(rcew_elev <0,np.nan, rcew_elev)

```

```python
plt.figure(figsize = (10,16))
plt.imshow(rcew_elev, cmap = 'terrain')
plt.colorbar()
plt.show
```

```python
f_elev, b_elev = np.histogram(rcew_elev[~np.isnan(rcew_elev)], bins=nbins)
print(f_elev)
print()
print()

print(b_elev)
```

```python
bin_centers = (b_elev[0:-1] + b_elev[1:])/2
bin_centers
```

```python

plt.figure(figsize=(12,10))
plt.plot(bin_centers, f_elev)
plt.xlabel('Elevation [m]')
plt.ylabel('Count [-]')
```

```python
hypso_rcew =1- np.cumsum((f_elev/np.sum(f_elev)))
hypso_rcew
```

```python
plt.figure(figsize=(12,12))
plt.plot(bin_centers,hypso_rcew,'ko:')
plt.xlabel('Elevation [m]')
plt.ylabel('Fraction of area above [-]')
```

```python
df = pd.DataFrame()
Rank=np.arange(35)+1

df['Rank'] = Rank
df['Fraction_Above']=hypso_rcew
df['Elevation']=bin_centers
print(df)

df2 = df.sort_values("Fraction_Above",ascending=True)
df2['Rank']=Rank
print()
print()
print()
print()
print(df2)
```

```python
z_33 = (1/3)
z_67 = (2/3)

ele_33 =np.interp(z_33,df2['Fraction_Above'],df2['Elevation'])
ele_67 =np.interp(z_67,df2['Fraction_Above'],df2['Elevation'])

print('The elevation corresponding to the 33rd percentile is {:.3f} m^3/s'.format(ele_33))
print('The elevation corresponding to the 67rd percentile is {:.3f} m^3/s'.format(ele_67))
```

```python
Bottom_third_ele = np.where(rcew_elev <ele_67,rcew_elev,np.nan)
Top_third_ele = np.where(rcew_elev > ele_33,rcew_elev,np.nan)
print(Bottom_third_ele)
print()
print()
print()
print()
print(Top_third_ele)

```

```python
print('The Digital Elevation Models (DEMs) of the Top one third elevations of the watersheds')
print()
print()
print()

plt.figure(figsize = (10,16))
plt.imshow(Top_third_ele, cmap = 'terrain')
plt.title("The Digital Elevation Models (DEMs) of the top one third elevations of the watershed",fontsize=20,color="green")
plt.colorbar()
plt.show
```

```python
print('The Digital Elevation Models (DEMs) of the bottom one third elevations of the watersheds')
print()
print()
print()

plt.figure(figsize = (10,16))
plt.imshow(Bottom_third_ele, cmap = 'terrain')
plt.title("The Digital Elevation Models (DEMs) of the bottom one third elevations of the watershed",fontsize=20,color="green")

plt.colorbar()
plt.show



```
